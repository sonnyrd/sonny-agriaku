### I. Import Library

```python
import pandas as pd
import os


pd.set_option('display.max_columns', None)
```

### II. Loading Dataset

```python
df = pd.DataFrame()
for i in os.listdir('dataset'):
    sensor = i.split('.')[0]
    raw_df = pd.read_csv(f'dataset/{i}',delimiter=';')
    raw_df['sensor'] = raw_df.apply(lambda x : sensor, axis=1)
    df = pd.concat([df,raw_df])
```

```python
df
```

```python
df['type'] = df[['1-Octanol','1-Propanol','2-Butanol','2-propanol','1-isobutanol']].idxmax(axis=1)
df.drop(columns=['1-Octanol','1-Propanol','2-Butanol','2-propanol','1-isobutanol'],inplace=True)
```
